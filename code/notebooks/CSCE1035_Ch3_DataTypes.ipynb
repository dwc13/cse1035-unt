### Named Tuples

Tuple subclass with named fields for better readability.

__Syntax:__ `Point = namedtuple('Point', ['x', 'y'])`

```python
from collections import namedtuple
Point = namedtuple("Point", ["x", "y", "z", "internsity"])
p1 = Point(10, 20, 15, 0.5)
print("Origin: ({}, {})".format(p1.x, p1.y))	            # Access by name
print("Intensity: {}\nDepth: {}".format(p1[3], p1[2]))  # Access by index
```

## 3.4 Set Basics

__Sets__: Unordered collection of unique elements. No indexing.

__Syntax:__ `s = {1, 2, 3}` or `s = set([1, 2, 3])`

```python
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print(s1 | s2)  # union
print(s1 & s2)  # intersection
print(s1 - s2)  # difference
print(s1 ^ s2)  # symmetric difference
```

## 3.5 Dictionary Basics

__Dictionaries__: Key-value pairs. Mutable, fast lookups using keys.

```python
student = {"name": "Tony", "age": 20, "debt": -1245.03}
student["major"] = "CS"
print(student)
print("{} is in ${} debt.".format(student['name'], student.get("debt")))
```

## 3.9 Binary Numbers

Binary is base-2. Use `bin()` and `int(x, 2)` for conversions.

```python
b = 0b10101010
print(b)  # 10
print(bin(170))
print(int('10101010', 2))
```

## 3.10 String Formatting

Use `f-strings` or `.format()` to embed variables in strings.

```python
student = {"name": "Tony", "age": 20, "debt": -1245.03}
student["major"] = "CS"
print(student)
print("{} is in ${} debt.".format(student['name'], student.get("debt")))
print(f"{student['name']} is in ${student.get('debt')} debt.")
print(f"{student['name']} is in ${student.get("debt")} debt.")
```
